```python
# Sample data: Average daily temperatures (in degrees Celsius) for a week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
temperatures = [22.5, 24.0, 23.5, 25.0, 26.5, 27.0, 26.0]


plt.plot(days,temperatures, label='Avg Temp',color='red')

#Adding title and labels

plt.title("Daily Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.ylim(20,30) # set the y - axis for better focus
plt.grid(True) # Grid for easier visualization

plt.legend()

plt.show()


```

```python
# To make 2 line chart in one

# Sample data: Monthly sales for two products over a year

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [15000, 18000, 17000, 16000, 20000, 21000, 23000, 25000, 24000, 26000, 27000, 30000]
product_B_sales = [12000, 13000, 13500, 14000, 14500, 15000, 16000, 18000, 19000, 19500, 20000, 22000]

# Plotting the data for product A

plt.plot(months,product_A_sales, label='Product A', color='blue',marker='o')

# Plotting the data for product B

plt.plot(months,product_B_sales, label='Product B',color='red',marker='s')

# Additing titles and labels

plt.title("Monthly sales comaparison of Product A and product B")
plt.xlabel("Months")
plt.ylabel("sales(USD)")
plt.ylim(10000,30000)
plt.grid(True)

plt.legend()

plt.show()
```

```python
# download historical data for apple
ticket='AAPL'
data=yf.download(ticket,start='2022-01-01',end='2023-01-01')
```

```python
# Plot the closing prices

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='AAPL Closing Price', color='blue')

# Add titles

plt.title('Apple Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

```python
# Load the data

california= fetch_california_housing()
```

```python
# Create a pandas dataframe for this dataset

df=pd.DataFrame(california.data)

df.head(5)
```

```python
df.columns= california.feature_names

df.columns
```

```python
# Using subplot

plt.figure(figsize=(25,25))

for i,col in enumerate(california.feature_names):
  plt.subplot(4,4,i+1)
  sns.regplot(x=col,y="AveRooms",data=df)
  plt.title("%s vs AveRooms" %(col))
```

```python
# Correlation - statistical relationship between 2 Variables
```

```python
data = {
    'Math': [90, 85, 88, 92, 85, 80, 78, 94, 90, 91],
    'Science': [85, 87, 85, 90, 82, 80, 77, 93, 89, 90],
    'English': [78, 83, 77, 85, 79, 75, 70, 88, 82, 86],
    'History': [72, 70, 65, 78, 74, 69, 68, 80, 75, 79],
}

# Convert to data frame

df=pd.DataFrame(data)

df.head()

# Calculate co realation matrix

correlation_matrix= df.corr()

# Create a heatmap

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5)

# Add a title
plt.title('Corelation heatmap')

# show the plot

plt.show()
```
