```python
# convert date to prper datetime format

df['Date']=pd.to_datetime(df['Date'], format = '%m/%d/%Y')
```

```python
# univariate analysis

df['Status'].value_counts()

# plot distribution of 'Amount' by sales

plt.figure(figsize=(10,5))
sns.histplot(df['Amount'])
plt.title('Distribution of sales Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# bivariate analysis

sales_per_day=df.groupby('Date')['Amount'].sum()

sales_per_day

# Plot sales trend over time

plt.figure(figsize=(10,5))
sns.lineplot(x=sales_per_day.index, y=sales_per_day.values)
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')

plt.show()
```

```python
# Sales by category
category_sales=df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

category_sales
```

```python
# Plot category wise sales

category_sales.plot(kind='bar', figsize=(12,6), color='orange')
plt.title('Total Sales by Product Category')
plt.ylabel('Amount')
plt.xlabel('Category')
plt.show()
```

```python
# Sales by state

state_sales= df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)

state_sales
```

```python
# plot state wise sales

state_sales.plot(kind='bar', figsize=(12,6), color='orange')
plt.title('Total Sales by State')
plt.ylabel('Amount')
plt.xlabel('State')
plt.show()
```
