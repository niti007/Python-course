```python
import requests
from bs4 import BeautifulSoup

# Books to Scrape scraper function defined here
def books_scraper(max_results=20):
    # Target URL
    url = "http://books.toscrape.com/catalogue/page-1.html"

    # Headers to mimic a real browser
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }

    # Send GET request
    response = requests.get(url, headers=headers)

    # Parse the response HTML using BeautifulSoup
    soup = BeautifulSoup(response.content, 'lxml')

    # Find all book product blocks
    products = soup.find_all("article", class_="product_pod")

    # List to store scraped data
    data = []

    # Loop through each product
    for product in products:
        # Get title from the 'title' attribute inside <a> tag
        title = product.h3.a['title']

        # Price is inside <p class="price_color">
        price = product.find("p", class_="price_color").text.strip()

        # Rating is a class on <p class="star-rating">
        rating_tag = product.find("p", class_="star-rating")
        rating = rating_tag['class'][1] if rating_tag else "No rating"

        # Availability info is on the next page (optional: skipped here)

        product_data = {
            "Title": title,
            "Price": price,
            "Rating": rating
        }

        data.append(product_data)

        # Break if max_results reached
        if len(data) >= max_results:
            break

    # Return the final list of products
    return data

# Example usage
if __name__ == "__main__":
    results = books_scraper(max_results=10)
    for i, item in enumerate(results, 1):
        print(f"{i}. {item['Title']} | {item['Price']} | Rating: {item['Rating']}")

```

```python
import schedule
import time

def job():
    print("Job chal raha hai")

# Har 5 second mein is job ko run karo
schedule.every(5).seconds.do(job)

while True:
    schedule.run_pending()
    time.sleep(1)
```

| Feature              | `schedule`                          | `time.sleep()`                  |
| -------------------- | ----------------------------------- | ------------------------------- |
| Purpose              | Task ko repeat karna at intervals   | Execution ko delay karna        |
| Time Control         | Seconds, minutes, days etc.         | Seconds-based delay             |
| Repeat Functionality | Haan, automatically repeat hota hai | Nahi, manually lagana padta hai |
| Example Use Case     | Background job scheduling           | Animation ya countdown delay    |
